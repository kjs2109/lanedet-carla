```python
import numpy as np

def binarize_array(arr):
    # 입력 배열이 0이 아닌 경우에는 1로 변환
    binary_arr = np.where(arr != 0, 1, 0)
    return binary_arr

# 예시 사용법
w, h, c = 4, 4, 3
example_array = np.random.randint(0, 10, (w, h, c))

print("Original Array:")
print(example_array)

binary_array = binarize_array(example_array)

print("Binarized Array:")
print(binary_array)
```

```python
src_image_root = '/root/works/lanedet/data/DSS/images' 

scenario_list = ['scenario_' + str(i) for i in range(1, 11)] 

# split by scenario 
scenario_group = filtered_df.groupby('scenario_name') 

for scenario_name, group in scenario_group: 
    
    image_paths = [] 
    for data_info in group.itertuples(): 
        image_name = data_info[1]
        image_path = os.path.join(src_image_root, scenario_name, image_name)  
        image_paths.append(image_path) 

        count_image = 0 
        with open(f'/root/works/lanedet/data/DSS/{scenario_name}.txt', 'w') as f: 
            for image_path in image_paths: 
                count_image += 1 
                f.write(image_path + '\n') 

    print(f'{scenario_name} is saved', count_image) 
    
```
